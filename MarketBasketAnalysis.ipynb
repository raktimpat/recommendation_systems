Let's take only UK transactions for simplification

```python
df = df[df['Country']=='United Kingdom']
df.shape
```

Convert it into wide form for more readablitiy

```python
data = (df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
data.head()
```

```python
data = (data > 0).astype(int)
data.head()
```

```python
data.shape
```

# Apriori Algorithm


>⚡ **min_support** : *ignore all supersets for itemsets that has freq < min_support*


**Note**: As we are recommending items that multiple people have bought in the past

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(data,
                                 min_support=0.028,
                                 use_colnames=True).sort_values("support", ascending=False).reset_index(drop=True)
```

```python
frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemsets_plus
```

**Association Rules :**



```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules = association_rules(frequent_itemsets_plus, metric='lift',
                                      min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
